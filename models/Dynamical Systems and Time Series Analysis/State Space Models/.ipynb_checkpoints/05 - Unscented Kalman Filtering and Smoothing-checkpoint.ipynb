```julia
using Distributions, LinearAlgebra, Plots, Random, LargeMessageCollider
Random.seed!(12345);
```

This example is due to Bayesian Filtering and Smoothing by Simo Sarkka (Example 5.3 Pendulum Tracking)

```julia
# Generate synthetic data

Δt = 0.01 # time gap
g = 10 # gravitational constant
q_c = 1 # spectral density of process noise
f(x::Array) = [x[1] + x[2]*Δt, x[2] - g*sin(x[1])*Δt] # nonlinear transition function
h(x::Array) = sin(x[1]) # nonlinear emission function
Q = [q_c*Δt^3/3 q_c*Δt^2/2; q_c*Δt^2/2 q_c*Δt] # transition noise covariance
R = 0.2 # emission noise variance
T = 500

# Generate data
α_real = [pi/2] # angle for visualization purpose
x_real = [[α_real[1],0]]
y = [h(x_real[1]) + sqrt(R)*randn()]
for t=2:T
    x_new = f(x_real[t-1]) # swinging in perfect conditions (no noise in process)
    push!(x_real, x_new)
    push!(α_real,x_new[1])
    y_new = h(x_real[t]) + sqrt(R)*randn() # imperfect measurements (noisy)
    push!(y, y_new)
end
```

```julia
plot(Δt * collect(1:500), α_real, lw=2, label="True process")
scatter!(Δt * collect(1:500), y, color=:black, markersize=2, label="Noisy measurements")
```

```julia
# Unscented Kalman Filtering

p1 = MvNormal(zeros(2),1) #prior

unscented = UT()

forward_estimates = [approximateMarginal!(unscented, h, y[1]+Normal(0,sqrt(R)), p1)]

# Recursive estimation
for t=2:T
    m_forw_, _ = forwardMessage(unscented, f, forward_estimates[t-1])
    m_forw = m_forw_ + MvNormal(zeros(2),Q)
    push!(forward_estimates, approximateMarginal!(unscented, h, y[t]+Normal(0,sqrt(R)), m_forw))
end
```

```julia
# Filtering Estimations
forward_estimates_α_mean = zeros(T)
forward_estimates_α_var = zeros(T)
for t=1:T
    forward_estimates_α_mean[t] = mean(forward_estimates[t])[1]
    forward_estimates_α_var[t] = cov(forward_estimates[t])[1,1]
end

plot(Δt * collect(1:500), α_real, lw=2, label="True process")
scatter!(Δt * collect(1:500), y, color=:black, markersize=2, label="Noisy measurements")
plot!(Δt * collect(1:500), forward_estimates_α_mean, ribbon=forward_estimates_α_var, alpha=0.2, lw=2, label="UKF")
```

```julia
# Unscented Kalman Smoothing
smooth_estimates = Array{MvNormal}(undef, T)
smooth_estimates[T] = forward_estimates[T]

joint_dists = Array{MvNormal}(undef, T-1)

for t=T-1:-1:1
    smooth_estimates[t], joint_dists[t] = transit(unscented, forward_estimates[t],smooth_estimates[t+1],f,inv(Q))
end
```

```julia
# Smoothing Estimations
smooth_estimates_α_mean = zeros(T)
smooth_estimates_α_var = zeros(T)
for t=1:T
    smooth_estimates_α_mean[t] = mean(smooth_estimates[t])[1]
    smooth_estimates_α_var[t] = cov(smooth_estimates[t])[1,1]
end

plot(Δt * collect(1:500), α_real, lw=2, label="True process")
scatter!(Δt * collect(1:500), y, color=:black, markersize=2, label="Noisy measurements")
plot!(Δt * collect(1:500), forward_estimates_α_mean, ribbon=forward_estimates_α_var, alpha=0.2, lw=2, label="EKF")
plot!(Δt * collect(1:500), smooth_estimates_α_mean, ribbon=smooth_estimates_α_var, alpha=0.2, lw=2, label="EKS")
```
